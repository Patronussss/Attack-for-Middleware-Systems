```python
import csv

def count_csv_column_items(file_path, column_name):
    """
    统计CSV文件中指定列的数据项数量
    
    参数:
    file_path: CSV文件的路径
    column_name: 要统计的列名
    
    返回:
    该列的数据项总数（包含空值）
    """
    try:
        # 初始化计数器
        count = 0
        # 打开CSV文件
        with open(file_path, 'r', encoding='utf-8') as csvfile:
            # 创建CSV读取器
            reader = csv.DictReader(csvfile)
            
            # 检查列名是否存在
            if column_name not in reader.fieldnames:
                print(f"错误：列名 '{column_name}' 不存在！")
                print(f"文件中可用的列名：{reader.fieldnames}")
                return 0
            
            # 遍历每一行，统计数据项
            for row in reader:
                # 即使单元格为空，也计数（空值也是一个数据项）
                count += 1
        
        return count
    
    except FileNotFoundError:
        print(f"错误：文件 '{file_path}' 未找到！")
        return 0
    except Exception as e:
        print(f"发生错误：{e}")
        return 0

# ------------------- 示例使用 -------------------
if __name__ == "__main__":
    # 替换为你的CSV文件路径和要统计的列名
    csv_file = "/media/data1/mcr/zl/zl/AttackForDataBlinder/dataset/usa_2_cipher.csv"  # 你的CSV文件路径
    target_column = "INCTOT"       # 要统计的列名
    
    # 调用函数统计
    item_count = count_csv_column_items(csv_file, target_column)
    print(f"列 '{target_column}' 共有 {item_count} 个数据项")
```
