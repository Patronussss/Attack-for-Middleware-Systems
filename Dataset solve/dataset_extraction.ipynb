```python
import csv
import re


def clean_column(csv_file_path, column_header):
    # 定义不能用作文件名的字符的正则表达式
    invalid_chars = r'[\\/*?:"<>|]'
    # 用于存储修改后的数据
    updated_rows = []

    try:
        # 打开 CSV 文件进行读取
        with open(csv_file_path, 'r', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            # 获取 CSV 文件的表头
            fieldnames = reader.fieldnames

            for row in reader:
                # 清理指定列的数据
                if column_header in row:
                    row[column_header] = re.sub(invalid_chars, '', row[column_header])
                updated_rows.append(row)

        # 打开 CSV 文件进行写入
        with open(csv_file_path, 'w', newline='', encoding='utf-8') as file:
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            # 写入表头
            writer.writeheader()
            # 写入修改后的数据
            writer.writerows(updated_rows)

        print("CSV 文件修改成功。")
    except FileNotFoundError:
        print(f"错误：未找到文件 {csv_file_path}。")
    except Exception as e:
        print(f"发生未知错误：{e}")


if __name__ == "__main__":
    csv_file = 'F:/Desktop/Supplementary experiments/Crime_plain.csv'
    header = 'Crm Cd Desc'
    clean_column(csv_file, header)    
```
