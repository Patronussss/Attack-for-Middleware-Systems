```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
def split_csv(file_path):
    try:
        # 读取 CSV 文件
        df = pd.read_csv(file_path)
        # 按 4:6 比例随机分割数据
        train_df, test_df = train_test_split(df, test_size=0.6, random_state=42)
        # 保存分割后的数据到新的 CSV 文件
        train_df.to_csv('Alzheimer_cipher.csv', index=False)
        test_df.to_csv('Alzheimer_plain.csv', index=False)
        print("文件已成功分割并保存为 train.csv 和 test.csv。")
    except FileNotFoundError:
        print("错误：文件未找到，请检查文件路径。")
    except Exception as e:
        print(f"发生未知错误：{e}")
```

```python
split_csv("E:/code/Searchable Encryption/dataset/Alzheimer_s_Disease_and_Healthy_Aging_Data.csv")    
```
